## Load Dataset

```python
import pandas as pd

# Load training data
train_sa_df = pd.read_csv('/content/train_sa_10000.csv')
train_en_df = pd.read_csv('/content/train_en_10000.csv')

# Load development data
dev_sa_df = pd.read_csv('/content/dev_sa_1000.csv')
dev_en_df = pd.read_csv('/content/dev_en_1000.csv')

# Load test data
test_sa_df = pd.read_csv('/content/test_sa_1000.csv')
test_en_df = pd.read_csv('/content/test_en_1000.csv')

print("Train Sanskrit data head:")
display(train_sa_df.head())

print("Train English data head:")
display(train_en_df.head())

print("Dev Sanskrit data head:")
display(dev_sa_df.head())

print("Dev English data head:")
display(dev_en_df.head())

print("Test Sanskrit data head:")
display(test_sa_df.head())

print("Test English data head:")
display(test_en_df.head())
```

## Merge Datasets

```python
# Merge training data
train_df = pd.merge(train_sa_df, train_en_df, on='Source_id', suffixes=('_sa', '_en'))

# Merge development data
dev_df = pd.merge(dev_sa_df, dev_en_df, on='Source_id', suffixes=('_sa', '_en'))

# Merge test data
test_df = pd.merge(test_sa_df, test_en_df, on='Source_id', suffixes=('_sa', '_en'))

print("Merged Train data head:")
display(train_df.head())

print("Merged Dev data head:")
display(dev_df.head())

print("Merged Test data head:")
display(test_df.head())
```

## Preprocess Text Data


```python
import re

def preprocess_sentence(sentence):

    sentence = sentence.lower()
    sentence = re.sub(r"([^0-9\u0900-\u097F\u0041-\u007A.,!?']+)", r" ", sentence)
    sentence = re.sub(r"([.,!?])", r" \1 ", sentence)
    sentence = re.sub(r"\s+", r" ", sentence).strip()
    sentence = '<start> ' + sentence + ' <end>'
    return sentence

# Apply preprocessing to Sanskrit and English sentences in all dataframes

# Train data
train_df['Sentence_sa_processed'] = train_df['Sentence_sa'].apply(preprocess_sentence)
train_df['Sentence_en_processed'] = train_df['Sentence_en'].apply(preprocess_sentence)

# Dev data
dev_df['Sentence_sa_processed'] = dev_df['Sentence_sa'].apply(preprocess_sentence)
dev_df['Sentence_en_processed'] = dev_df['Sentence_en'].apply(preprocess_sentence)

# Test data
test_df['Sentence_sa_processed'] = test_df['Sentence_sa'].apply(preprocess_sentence)
test_df['Sentence_en_processed'] = test_df['Sentence_en'].apply(preprocess_sentence)

print("Processed Train data head:")
display(train_df[['Sentence_sa_processed', 'Sentence_en_processed']].head())

print("Processed Dev data head:")
display(dev_df[['Sentence_sa_processed', 'Sentence_en_processed']].head())

print("Processed Test data head:")
display(test_df[['Sentence_sa_processed', 'Sentence_en_processed']].head())
```

## Create Vocabulary (Tokenization and Indexing)

This step involves creating a word-to-index mapping for both Sanskrit and English sentences using the processed training data.

```python
import tensorflow as tf

# Create Sanskrit tokenizer
sa_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='', oov_token='<unk>')
sa_tokenizer.fit_on_texts(train_df['Sentence_sa_processed'])

# Create English tokenizer
en_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='', oov_token='<unk>')
en_tokenizer.fit_on_texts(train_df['Sentence_en_processed'])

# Convert text to sequences of integers
train_sa_sequences = sa_tokenizer.texts_to_sequences(train_df['Sentence_sa_processed'])
train_en_sequences = en_tokenizer.texts_to_sequences(train_df['Sentence_en_processed'])

dev_sa_sequences = sa_tokenizer.texts_to_sequences(dev_df['Sentence_sa_processed'])
dev_en_sequences = en_tokenizer.texts_to_sequences(dev_df['Sentence_en_processed'])

test_sa_sequences = sa_tokenizer.texts_to_sequences(test_df['Sentence_sa_processed'])
test_en_sequences = en_tokenizer.texts_to_sequences(test_df['Sentence_en_processed'])

# Calculate vocabulary sizes
SA_VOCAB_SIZE = len(sa_tokenizer.word_index) + 1  # +1 for padding
EN_VOCAB_SIZE = len(en_tokenizer.word_index) + 1  # +1 for padding

print(f"Sanskrit vocabulary size: {SA_VOCAB_SIZE}")
print(f"English vocabulary size: {EN_VOCAB_SIZE}")

# Display a few examples of tokenized sequences
print("\nExample Sanskrit processed sentence and its sequence:")
print(train_df['Sentence_sa_processed'].iloc[0])
print(train_sa_sequences[0])

print("\nExample English processed sentence and its sequence:")
print(train_df['Sentence_en_processed'].iloc[0])
print(train_en_sequences[0])
```

## Pad Sequences

This step pads all sequences to ensure they have the same length, which is required for batching in neural networks.

```python
# Determine maximum sequence length for Sanskrit and English from training data
MAX_SA_LEN = max(len(seq) for seq in train_sa_sequences)
MAX_EN_LEN = max(len(seq) for seq in train_en_sequences)

print(f"Maximum Sanskrit sequence length: {MAX_SA_LEN}")
print(f"Maximum English sequence length: {MAX_EN_LEN}")

# Pad all sequences
train_sa_padded = tf.keras.preprocessing.sequence.pad_sequences(train_sa_sequences, maxlen=MAX_SA_LEN, padding='post')
train_en_padded = tf.keras.preprocessing.sequence.pad_sequences(train_en_sequences, maxlen=MAX_EN_LEN, padding='post')

dev_sa_padded = tf.keras.preprocessing.sequence.pad_sequences(dev_sa_sequences, maxlen=MAX_SA_LEN, padding='post')
dev_en_padded = tf.keras.preprocessing.sequence.pad_sequences(dev_en_sequences, maxlen=MAX_EN_LEN, padding='post')

test_sa_padded = tf.keras.preprocessing.sequence.pad_sequences(test_sa_sequences, maxlen=MAX_SA_LEN, padding='post')
test_en_padded = tf.keras.preprocessing.sequence.pad_sequences(test_en_sequences, maxlen=MAX_EN_LEN, padding='post')

print("\nExample of padded Sanskrit sequence (first training example):")
print(train_sa_padded[0])

print("\nExample of padded English sequence (first training example):")
print(train_en_padded[0])
```

## Build the Encoder-Decoder Model

We will implement a custom sequence-to-sequence model with an Encoder-Decoder architecture using GRU layers and Bahdanau attention.

```python
import tensorflow as tf

class Encoder(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, enc_units, batch_sz):
        super(Encoder, self).__init__()
        self.batch_sz = batch_sz
        self.enc_units = enc_units
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(self.enc_units,
                                       return_sequences=True,
                                       return_state=True,
                                       recurrent_initializer='glorot_uniform')

    def call(self, x, hidden):
        x = self.embedding(x)
        output, state = self.gru(x, initial_state=hidden)
        return output, state

    def initialize_hidden_state(self):
        return tf.zeros((self.batch_sz, self.enc_units))

# Define model parameters
EMBEDDING_DIM = 256
UNITS = 1024
BATCH_SIZE = 64 # You might need to adjust this based on your GPU memory

# Instantiate the Encoder
encoder = Encoder(SA_VOCAB_SIZE, EMBEDDING_DIM, UNITS, BATCH_SIZE)

```

```python
class BahdanauAttention(tf.keras.layers.Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = tf.keras.layers.Dense(units)
        self.W2 = tf.keras.layers.Dense(units)
        self.V = tf.keras.layers.Dense(1)

    def call(self, query, values):

        hidden_with_time_axis = tf.expand_dims(query, 1)


        score = self.V(tf.nn.tanh(self.W1(values) + self.W2(hidden_with_time_axis)))

        # attention_weights shape == (batch_size, max_len, 1)
        attention_weights = tf.nn.softmax(score, axis=1)

        # context_vector shape == (batch_size, hidden_size)
        context_vector = attention_weights * values
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector, attention_weights

# Instantiate the Bahdanau Attention layer
attention_layer = BahdanauAttention(UNITS)

```

```python
class Decoder(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, dec_units, batch_sz):
        super(Decoder, self).__init__()
        self.batch_sz = batch_sz
        self.dec_units = dec_units
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(self.dec_units,
                                       return_sequences=True,
                                       return_state=True,
                                       recurrent_initializer='glorot_uniform')
        self.fc = tf.keras.layers.Dense(vocab_size)

        # Used for attention
        self.attention = BahdanauAttention(self.dec_units)

    def call(self, x, hidden, enc_output):
        # enc_output shape == (batch_size, max_length, hidden_size)
        context_vector, attention_weights = self.attention(hidden, enc_output)

        # x shape after passing through embedding == (batch_size, 1, embedding_dim)
        x = self.embedding(x)

        # x shape after concatenation == (batch_size, 1, embedding_dim + hidden_size)
        x = tf.concat([tf.expand_dims(context_vector, 1), x], axis=-1)

        # passing the concatenated vector to the GRU
        output, state = self.gru(x)

        # output shape == (batch_size * 1, hidden_size)
        output = tf.reshape(output, (-1, output.shape[2]))

        # output shape == (batch_size, vocab)
        x = self.fc(output)

        return x, state, attention_weights

# Instantiate the Decoder
decoder = Decoder(EN_VOCAB_SIZE, EMBEDDING_DIM, UNITS, BATCH_SIZE)

```

## Configure Training

This section sets up the optimizer, loss function, and prepares the `tf.data.Dataset` objects for training.

```python
import os

optimizer = tf.keras.optimizers.Adam()
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(
    from_logits=True, reduction='none')

def loss_function(real, pred):
    mask = tf.math.logical_not(tf.math.equal(real, 0))
    loss_ = loss_object(real, pred)
    mask = tf.cast(mask, dtype=loss_.dtype)
    loss_ *= mask
    return tf.reduce_mean(loss_)

# Create tf.data.Dataset objects
# Target input for decoder should exclude the <end> token
# Target real for loss calculation should exclude the <start> token

train_dataset = tf.data.Dataset.from_tensor_slices(
    (train_sa_padded, train_en_padded[:, :-1], train_en_padded[:, 1:]))

train_dataset = train_dataset.shuffle(len(train_sa_padded)).batch(BATCH_SIZE, drop_remainder=True)

dev_dataset = tf.data.Dataset.from_tensor_slices(
    (dev_sa_padded, dev_en_padded[:, :-1], dev_en_padded[:, 1:]))

dev_dataset = dev_dataset.batch(BATCH_SIZE, drop_remainder=True)

# Configure checkpointing
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "ckpt")
checkpoint = tf.train.Checkpoint(optimizer=optimizer,
                                 encoder=encoder,
                                 decoder=decoder)
```

## Training Loop

This section defines the training step, evaluation step, and the main training loop.

```python
import time
import numpy as np

@tf.function
def train_step(inp, targ_inp, targ_real, enc_hidden):
    loss = 0

    with tf.GradientTape() as tape:
        enc_output, enc_hidden = encoder(inp, enc_hidden)

        dec_hidden = enc_hidden

        for t in range(targ_inp.shape[1]):
            # passing enc_output to the decoder
            predictions, dec_hidden, _ = decoder(tf.expand_dims(targ_inp[:, t], 1), dec_hidden, enc_output)

            loss += loss_function(targ_real[:, t], predictions)

    batch_loss = (loss / int(targ_inp.shape[1]))

    variables = encoder.trainable_variables + decoder.trainable_variables
    gradients = tape.gradient(loss, variables)
    optimizer.apply_gradients(zip(gradients, variables))

    return batch_loss

@tf.function
def evaluate_step(inp, targ_inp, targ_real, enc_hidden):
    loss = 0

    enc_output, enc_hidden = encoder(inp, enc_hidden)
    dec_hidden = enc_hidden

    for t in range(targ_inp.shape[1]):
        predictions, dec_hidden, _ = decoder(tf.expand_dims(targ_inp[:, t], 1), dec_hidden, enc_output)
        loss += loss_function(targ_real[:, t], predictions)

    batch_loss = (loss / int(targ_inp.shape[1]))
    return batch_loss


EPOCHS = 10

for epoch in range(EPOCHS):
    start = time.time()

    enc_hidden = encoder.initialize_hidden_state()
    total_train_loss = 0
    total_dev_loss = 0

    for (batch, (inp, targ_inp, targ_real)) in enumerate(train_dataset):
        batch_loss = train_step(inp, targ_inp, targ_real, enc_hidden)
        total_train_loss += batch_loss

        if batch % 100 == 0:
            print(f'Epoch {epoch+1} Batch {batch} Train Loss {batch_loss.numpy():.4f}')

    for (batch, (inp, targ_inp, targ_real)) in enumerate(dev_dataset):
        batch_loss = evaluate_step(inp, targ_inp, targ_real, enc_hidden)
        total_dev_loss += batch_loss

    avg_train_loss = total_train_loss / len(train_dataset)
    avg_dev_loss = total_dev_loss / len(dev_dataset)

    print(f'Epoch {epoch+1} Train Loss {avg_train_loss:.4f} Dev Loss {avg_dev_loss:.4f}')

    # saving (checkpoint) the model every 2 epochs
    if (epoch + 1) % 2 == 0:
        checkpoint.save(file_prefix=checkpoint_prefix)

    print(f'Time taken for 1 epoch {time.time() - start:.2f} sec\n')
```

## Translate Function

This function will take a Sanskrit sentence as input and return its English translation using the trained model.

```python
def translate(sentence):
    # Preprocess the input sentence
    sentence = preprocess_sentence(sentence)

    # Convert to sequence of integers
    inputs = [sa_tokenizer.word_index[i] for i in sentence.split(' ')]
    inputs = tf.keras.preprocessing.sequence.pad_sequences([inputs],
                                                           maxlen=MAX_SA_LEN,
                                                           padding='post')
    inputs = tf.convert_to_tensor(inputs)

    result = []

    # Initialize encoder hidden state
    enc_hidden = encoder.initialize_hidden_state()
    enc_out, enc_hidden = encoder(inputs, enc_hidden)

    # Initialize decoder hidden state with encoder's last hidden state
    dec_hidden = enc_hidden

    # Start with <start> token for decoder input
    dec_input = tf.expand_dims([en_tokenizer.word_index['<start>']], 0)

    for t in range(MAX_EN_LEN):
        predictions, dec_hidden, attention_weights = decoder(dec_input,
                                                             dec_hidden,
                                                             enc_out)

        # Get the predicted word ID
        predicted_id = tf.argmax(predictions[0]).numpy()
        result.append(en_tokenizer.index_word[predicted_id])

        # If <end> token is predicted, stop translation
        if en_tokenizer.index_word[predicted_id] == '<end>':
            return ' '.join(result[:-1]), sentence # Exclude <end> token

        # The predicted ID is fed back into the model as the next input
        dec_input = tf.expand_dims([predicted_id], 0)

    return ' '.join(result), sentence
```

## Generate Translations for Test Set and Calculate Inference Time

```python
import time

predictions_list = []
source_ids = test_df['Source_id'].tolist()
sanskrit_sentences = test_df['Sentence_sa'].tolist()

start_inference_time = time.time()

for i, sa_sentence in enumerate(sanskrit_sentences):
    translated_text, _ = translate(sa_sentence)
    predictions_list.append((source_ids[i], translated_text))

end_inference_time = time.time()
total_inference_time = end_inference_time - start_inference_time

print(f"Total inference time for the test set: {total_inference_time:.2f} seconds")

# Create a DataFrame for submission
submission_df = pd.DataFrame(predictions_list, columns=['Source_id', 'Sentence_en'])

print("\nGenerated Submission DataFrame head:")
display(submission_df.head())

# Save to submission.csv
submission_df.to_csv('submission.csv', index=False, encoding='utf-8')
print("Submission file 'submission.csv' created successfully.")
```
